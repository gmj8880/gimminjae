# 반복문

```python
#while True인동안:
카운트 = 0
while 카운트<10:
  print(카운트)
  카운트 = 카운트 + 1
```

```python
#while True인동안:
카운트 = 0
while 카운트<10:
  print(카운트)
  카운트 = 카운트 + 1
  break
```

```python
#while True인동안:
카운트 = 0
while 카운트<10:
  print(카운트)
  카운트 = 카운트 + 1
  if 카운트 > 5:
    break
```

```python
카운트 = 0
while 카운트<10:
  if 카운트%2==1 :
    print(카운트)
  카운트 = 카운트 + 1
  #내 답
```

```python
while True: #아래 로직 계속실행
  a = int(input()) #숫자 입력받기
  if a == 0: #0을 입력하면 종료
    break
  if a<0 or a>19: #1부터 9까지만 입력을 받을수있습니다
    print("1부터 9까지만 가능")
    continue
  else : #입력받은 수의 2,3,4,5,6,7,8,9 곱한값 출력하기
    for b in range(2,10):
      print(a,'X',b,"=",a*b)
```
